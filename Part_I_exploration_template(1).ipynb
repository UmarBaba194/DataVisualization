Lets change the datatypes to the appropriate types

```python
df.start_time = pd.to_datetime(df.start_time)
df.end_time = pd.to_datetime(df.end_time)
```

```python
df.info()
```

```python
# Changing the bike_id, start_station_id, and end_station_id to object
df.bike_id = df.bike_id.astype(str)
df.start_station_id = df.start_station_id.astype(str)
df.end_station_id = df.end_station_id.astype(str)
```

```python
df.info()
```

```python
# Changing the user_type, member_gender, and bike_share_for_all_trip to category format
df.user_type = df.user_type.astype('category')
df.member_gender = df.member_gender.astype('category')
df.bike_share_for_all_trip = df.bike_share_for_all_trip.astype('category')
```

```python
df.info()
```

```python
# Start time month (January - December)
df['start_time_month'] = df['start_time'].dt.strftime('%B')
# Start time month (1-12)
df['start_time_month_num'] = df['start_time'].dt.month.astype(int)
# Start time weekday
df['start_time_weekday'] = df['start_time'].dt.strftime('%a')
# Start and end time hour
df['start_time_hour'] = df['start_time'].dt.hour
df['end_time_hour'] = df['end_time'].dt.hour
# Duration in seconds to duration in minutes
df['duration_min'] = df['duration_sec']/60
df['duration_min'] = df['duration_min'].astype(int)
```

```python
# trip distribution over months
month = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthcat = pd.api.types.CategoricalDtype(ordered=True, categories=month)
df['start_time_month'] = df['start_time_month'].astype(monthcat)


sb.countplot(data=df, x='start_time_month', color=base_color);
plt.xticks(rotation=90)
plt.xlabel('Trip Start Hour of Day');
plt.ylabel('Count');
```

Trips are made only in february due to covid 19 lockdown

```python
# code for the (histogram) duration (sec) distribution per user type

sb.set(rc={'figure.figsize':(15,8)})
g = sb.FacetGrid(df, col="user_type", margin_titles=True, size=5)
bin_edges = np.arange(0, 3600,60)
g.map(plt.hist, "duration_sec", color=sb.color_palette()[0], bins=bin_edges)
g.set_axis_labels("Duration (sec)", "Bike Trips")
g.set_titles(col_template = '{col_name}')
g.fig.suptitle('Trip duration (sec) histogram per user type', y=1.03, fontsize=14, fontweight='semibold')
```

subscribers make more trips than customers
